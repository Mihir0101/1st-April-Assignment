## Q4. What is the ROC curve, and how is it used to evaluate the performance of the logistic regression model?

- > The ROC curve is a graphical representation of the performance of a binary classification model across different classification thresholds. 

* How is is used to evaluate performance?

Train your logistic regression model and obtain predicted probabilities for each instance.

The ROC curve is generated by varying the classification threshold from 0 to 1. At each threshold, instances with predicted probabilities greater than or equal to the threshold are classified as positive, and others as negative.

Calculate the True Positive Rate (Sensitivity) and False Positive Rate for each threshold setting.

Plot the TPR against the FPR for different threshold settings, resulting in the ROC curve.

The AUC-ROC is a summary metric that quantifies the overall performance of the model. A higher AUC-ROC value (closer to 1) indicates better discrimination ability.
